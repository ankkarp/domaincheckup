```python
from time import sleep
import socket, threading

from pythonping import ping


class DomainCheckup:
    def __init__(self, timeout=2, count=4, interval=0, max_threads=100, verbose=False):
        self.timeout = timeout
        self.count = count
        self.interval = interval
        self.verbose = verbose
        self.max_threads = max_threads
        self.results = {}
        
    def ping_domain(self, target):
        self.results[target]['ping'] = ping(target,
            timeout=self.timeout,
            count=self.count,
            interval=self.interval,
            verbose=self.verbose
            ).__dict__

    def check_port_availability(self, target, port):
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM) # TCP
        #sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM) # UDP
        socket.setdefaulttimeout(self.timeout) # seconds (float)
        result = sock.connect_ex((target,port))
        self.results[target]['ports'][str(port)]['status'] = "OPEN" if result == 0 else "CLOSED"
        sock.close()
            
    def __call__(self, checklist):
        for target, ports in checklist.items():
            self.results[target] = {"ports": {str(port): {} for port in ports}}
            threading.Thread(target=self.ping_domain, args=[str(target)]).start()
            for port in ports: 
                threading.Thread(target=self.check_port_availability, args=[str(target), port]).start()
                threading.active_count()
                while threading.active_count() > self.max_threads:
                    sleep(1)
```

```python
import time
import asyncio

from collections.abc import Iterable

async def wait_host_port(host, ports=[None], timeout=2, count=4, interval=0):
    """Repeatedly try if a port on a host is open until duration seconds passed
    
    Parameters
    ----------
    host : str
        host ip address or hostname
    port : int
        port number
    duration : int, optional
        Total duration in seconds to wait, by default 10
    delay : int, optional
        delay in seconds between each try, by default 2
    
    Returns
    -------
    awaitable bool
    """
    if not isinstance(ports, Iterable):
        ports = [ports]
    res = []
    for p in ports:
            try:
                for i in range(count):
                    t0 = time.time()
                    _reader, writer = await asyncio.wait_for(asyncio.open_connection(host, p), timeout=timeout)
                    t1 = time.time()
                    print(t0, t1, t1-t0)
                    writer.close()
                    await writer.wait_closed()
                    res.append(True)
            except:
                if interval:
                    await asyncio.sleep(interval)
                    res.append(False)
    return res
```
